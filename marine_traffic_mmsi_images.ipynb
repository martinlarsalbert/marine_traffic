# Retrieve images of ships from Marine Traffic

```python
from selenium import webdriver as webdriver
import re
import urllib
```

```python
browser = webdriver.Chrome(executable_path="/Users/martinalexandersson/Dev/Selenium-Webscraping-Example/chromedriver")
```

```python
browser.get("""https://www.marinetraffic.com""")
```

```python
browser.get("""https://www.marinetraffic.com/en/ais/details/ships/shipid:371681/""")
```

```python
element = browser.find_element_by_id('big-image')
```

```python
href = element.get_attribute("href")
href
```

```python
result = re.search('#(.*)',href)
picture_id = result.groups(1)[0]
picture_id
```

```python
picture_href = 'https://photos.marinetraffic.com/ais/showphoto.aspx?photoid=%s' % picture_id
```

```python
picture_href
```

```python
browser.get(picture_href)
```

```python
browser.save_screenshot("image.png")
```
